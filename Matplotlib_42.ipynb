## 1. 🎯 Introduction to Matplotlib

### What is Matplotlib?

**Matplotlib** is a Python library for creating **beautiful, professional visualizations** (charts, graphs, plots, and images). Think of it as the **"painter"** of Python data science — it turns numbers into pictures!

Matplotlib helps you:
- 📈 Create line graphs, bar charts, histograms, scatter plots
- 🎨 Customize colors, styles, sizes, and fonts
- 📊 Add titles, labels, legends, and grids
- 🖼️ Save plots as images (PNG, PDF, SVG)
- 🎭 Create publication-quality scientific figures
- 🌐 Make interactive plots (with extra tools)
- 📐 Plot mathematical functions and data analysis results

### Why is it Called "Matplotlib"?

The name **"Matplotlib"** stands for **"MATLAB Plotting Library"**. It was created by a scientist who loved MATLAB's plotting features but wanted them in Python. MATLAB is expensive software used in engineering, so Matplotlib brought similar plotting power to free, open-source Python.

### Why Was Matplotlib Created?

Before Matplotlib, Python didn't have great visualization tools. Scientists and engineers needed a way to:
- ✅ Plot data and scientific results
- ✅ Create publication-quality figures for research papers
- ✅ Visualize large datasets easily
- ✅ Have an interface similar to MATLAB (so MATLAB users could switch to Python)

### Problems Matplotlib Solves

| Problem | How Matplotlib Solves It |
|---------|-------------------|
| Numbers are hard to understand | Converting data to visual plots makes patterns obvious |
| Comparing values | Bar/line charts make comparisons easy |
| Showing trends over time | Line plots clearly show how things change |
| Displaying data distributions | Histograms and box plots show patterns |
| Making professional reports | High-quality, customizable plots for presentations |
| Exploring data | Quick plotting helps spot issues and patterns |

### Why John D. Hunter Created Matplotlib

John D. Hunter was a neurobiologist doing brain signal (EEG) research. He needed to plot his research data but found Python's visualization tools inadequate. Being inspired by MATLAB's plotting capabilities, he created Matplotlib to give Python the same powerful plotting features. Over time, it became the most popular plotting library in Python!

### Major Milestones & Versions

| Year | Version | Major Features |
|------|---------|------------------|
| 2003 | 0.1 | Initial release with basic plot, imshow, contour |
| 2007 | 0.82 | Major improvements, Agg backend introduced |
| 2009 | 0.99 | Better performance, more plot types |
| 2012 | 1.1 | Three-dimensional (3D) plotting support added |
| 2016 | 2.0 | Major redesign - better defaults, cleaner API |
| 2018 | 3.0 | Python 2 support dropped, focus on Python 3 |
| 2020 | 3.3+ | Constrained layout, better performance |
| 2023+ | 3.6+ | Continued improvements and new features |

### How Matplotlib Fits Into the Ecosystem

Today, Matplotlib is the **foundation** of Python visualization:
- 🐼 **Pandas** - has `.plot()` built on top of Matplotlib
- 📈 **Seaborn** - statistical graphics library that wraps Matplotlib
- 🤖 **Scikit-learn** - uses Matplotlib for visualization functions
- 📊 **Plotly** - similar but more interactive (different tool)
- 🎯 **Data scientists everywhere** use Matplotlib for exploratory plots

**Fun fact:** Almost every other Python plotting library either uses Matplotlib underneath or was inspired by it!

## 4. 🧠 Basic Concepts

### Important Terminology

| Term | Meaning |
|------|---------|
| **Figure** | The entire plot window/canvas (like a blank piece of paper) |
| **Axes** | The actual area where data is drawn (the "plot" you see) |
| **Plot** | A specific type of visualization (line plot, bar plot, etc.) |
| **Artist** | Any object you can see on the plot (lines, text, markers, etc.) |
| **Backend** | The underlying system that actually draws the plot (e.g., "Agg" for saving, "TkAgg" for GUI windows) |
| **DPI** | Dots Per Inch — resolution quality of the saved image |
| **Subplot** | Multiple plots arranged in a grid on one figure |
| **Legend** | Box showing what different colors/styles represent |
| **Annotation** | Text labels and arrows pointing to specific data points |

### Core Components

1. **`pyplot`** (`plt`) — the main interface for creating and customizing plots
2. **`Figure`** — the overall container for everything
3. **`Axes`** — the plot area where data is actually drawn
4. **Plot types** — line, scatter, bar, histogram, box, etc.

### Syntax and Structure

A simple Matplotlib plot has this structure:

```python
import matplotlib.pyplot as plt

# Create a figure and axes (or just use defaults)
plt.figure()

# Plot some data
plt.plot([1, 2, 3], [1, 2, 3])

# Add labels and title
plt.title("Title")
plt.xlabel("X Label")
plt.ylabel("Y Label")

# Display the plot
plt.show()
```

### How Matplotlib Works Internally (Explained Simply)

Imagine you're drawing a picture with tools:
1. **Figure** = the blank canvas/piece of paper
2. **Axes** = the frame that defines the area where you draw
3. **Plot functions** = your drawing tools (pencil for lines, eraser, paint bucket, etc.)
4. **Show** = displaying the finished drawing

When you use Matplotlib, you're essentially:
1. Creating a blank canvas (Figure)
2. Adding a plotting area (Axes)
3. Using tools to draw (plot functions like `.plot()`, `.bar()`, etc.)
4. Adding labels and decorations (title, labels, legend)
5. Displaying it (`.show()`)

## 5. 🛠️ Functions and Features

### 5.1 Basic Plotting Functions

| Function | What it Does | Best For |
|----------|--------------|----------|
| `plt.plot()` | Creates a line plot | Showing trends, continuous data |
| `plt.scatter()` | Creates a scatter plot (dots) | Showing relationships between two variables |
| `plt.bar()` | Creates a bar chart | Comparing categories |
| `plt.barh()` | Creates a horizontal bar chart | Long category names |
| `plt.hist()` | Creates a histogram | Showing distribution of data |
| `plt.pie()` | Creates a pie chart | Showing parts of a whole (percentages) |
| `plt.boxplot()` | Creates a box plot | Showing data spread and outliers |
| `plt.fill_between()` | Fills area between lines | Showing confidence intervals |

### 5.2 Customization Functions

| Function | What it Does |
|----------|--------------|
| `plt.title()` | Adds title to the plot |
| `plt.xlabel()` / `plt.ylabel()` | Adds labels to axes |
| `plt.legend()` | Shows legend (what each line/color means) |
| `plt.grid()` | Adds grid lines |
| `plt.xlim()` / `plt.ylim()` | Sets the limits of axes |
| `plt.xticks()` / `plt.yticks()` | Customizes tick marks and labels |
| `plt.style.use()` | Changes the overall "theme" of the plot |
| `plt.annotate()` | Adds text and arrow pointing to a point |
| `plt.text()` | Adds text anywhere on the plot |

### 5.3 Figure and Axes Functions

| Function | What it Does |
|----------|--------------|
| `plt.figure()` | Creates a new figure |
| `plt.subplots()` | Creates multiple plots in a grid |
| `plt.subplot()` | Creates a subplot within a grid |
| `plt.tight_layout()` | Automatically spaces subplots nicely |
| `plt.savefig()` | Saves plot to a file (PNG, PDF, SVG, etc.) |
| `plt.show()` | Displays the plot |
| `plt.clf()` | Clears the current figure |
| `plt.close()` | Closes a figure window |

### 5.4 Styling Functions

| Function | What it Does |
|----------|--------------|
| `color=` parameter | Set line/marker color (e.g., 'red', '#FF0000') |
| `linestyle=` parameter | Set line style (solid, dashed, dotted, etc.) |
| `linewidth=` / `lw=` | Set thickness of lines |
| `marker=` parameter | Add markers to points ('o', 's', '^', etc.) |
| `markersize=` / `ms=` | Set size of markers |
| `alpha=` parameter | Set transparency (0=invisible, 1=opaque) |
| `label=` parameter | Name for legend entry |

### 5.5 Special Plot Types

| Function | Purpose |
|----------|---------|
| `plt.imshow()` | Display images or 2D data as colors |
| `plt.contour()` | Contour plot (like elevation map) |
| `plt.heatmap()` (in Seaborn) | Color-coded 2D array |
| `plt.streamplot()` | Vector field visualization |
| `plt.quiver()` | Arrows showing vectors |

### Example 1: Your Very First Plot (Simple Line Plot)

````python
import matplotlib.pyplot as plt

# Create data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the plot
plt.plot(x, y)
plt.title("My First Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
````

### Example 2: Customizing the Line Plot (Colors, Styles, Markers)

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot with customizations
plt.plot(x, y, color='red', linewidth=2, linestyle='--', marker='o', markersize=8, label='Data')
plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # Shows the label
plt.grid(True)  # Adds grid lines
plt.show()
````

### Example 3: Multiple Lines on One Plot

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 2, 3, 5]
y2 = [2, 5, 3, 4, 6]
y3 = [3, 1, 4, 2, 5]

# Plot multiple lines
plt.plot(x, y1, label='Line 1', marker='o')
plt.plot(x, y2, label='Line 2', marker='s')
plt.plot(x, y3, label='Line 3', marker='^')

plt.title("Multiple Lines")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
````

### Example 4: Scatter Plot (Dots Only, No Lines)

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [2, 4, 5, 4, 6, 7]

# Create scatter plot
plt.scatter(x, y, s=100, color='blue', alpha=0.6)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
````

### Example 7: Bar Chart with Multiple Groups

````python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Q1', 'Q2', 'Q3', 'Q4']
sales_2023 = [10, 15, 12, 20]
sales_2024 = [12, 18, 15, 25]

x = np.arange(len(categories))
width = 0.35

plt.bar(x - width/2, sales_2023, width, label='2023', color='blue')
plt.bar(x + width/2, sales_2024, width, label='2024', color='orange')

plt.title("Sales Comparison: 2023 vs 2024")
plt.xlabel("Quarters")
plt.ylabel("Sales (in millions)")
plt.xticks(x, categories)
plt.legend()
plt.show()
````

### Example 9: Histogram (Distribution of Data)

````python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.normal(100, 15, 1000)  # mean=100, std=15, 1000 samples

plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Distribution of Test Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.axvline(np.mean(data), color='red', linestyle='--', linewidth=2, label=f'Mean: {np.mean(data):.1f}')
plt.legend()
plt.show()
````

### Example 13: Subplots (Multiple Plots in One Figure)

````python
import matplotlib.pyplot as plt
import numpy as np

# Create 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1: Line plot
axes[0, 0].plot([1, 2, 3, 4], [1, 4, 2, 3])
axes[0, 0].set_title("Line Plot")

# Plot 2: Scatter plot
axes[0, 1].scatter([1, 2, 3, 4], [1, 4, 2, 3])
axes[0, 1].set_title("Scatter Plot")

# Plot 3: Bar chart
axes[1, 0].bar(['A', 'B', 'C', 'D'], [10, 24, 36, 18])
axes[1, 0].set_title("Bar Chart")

# Plot 4: Histogram
axes[1, 1].hist(np.random.normal(100, 15, 1000), bins=30)
axes[1, 1].set_title("Histogram")

plt.tight_layout()  # Prevents overlap
plt.show()
````

### Example 16: Logarithmic Scale

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 100)
y = np.exp(x/50)  # Exponential growth

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Linear scale
ax1.plot(x, y, color='blue')
ax1.set_title("Linear Scale")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")

# Logarithmic scale
ax2.semilogy(x, y, color='red')  # Logarithmic y-axis
ax2.set_title("Logarithmic Y-axis")
ax2.set_xlabel("X")
ax2.set_ylabel("Y (log scale)")

plt.tight_layout()
plt.show()
````

### Example 17: Annotating Points on a Plot

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 2, 3, 5]

plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', markersize=8)

# Annotate specific points
for i, (xi, yi) in enumerate(zip(x, y)):
    plt.annotate(f'Point {i+1}', xy=(xi, yi), xytext=(5, 5),
                textcoords='offset points', fontsize=9,
                bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7),
                arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))

plt.title("Annotated Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
````

### Example 18: Plotting Mathematical Functions

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2*np.pi, 2*np.pi, 200)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label='sin(x)', linewidth=2)
plt.plot(x, y_cos, label='cos(x)', linewidth=2)

plt.title("Trigonometric Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.axhline(y=0, color='k', linewidth=0.5)  # Add horizontal line at y=0
plt.axvline(x=0, color='k', linewidth=0.5)  # Add vertical line at x=0
plt.show()
````

### Example 19: Stacked Bar Chart

````python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Q1', 'Q2', 'Q3', 'Q4']
product_a = [10, 15, 12, 20]
product_b = [8, 12, 10, 15]
product_c = [6, 9, 8, 12]

x = np.arange(len(categories))
width = 0.6

plt.bar(x, product_a, width, label='Product A')
plt.bar(x, product_b, width, bottom=product_a, label='Product B')
plt.bar(x, product_c, width, bottom=np.array(product_a)+np.array(product_b), label='Product C')

plt.title("Stacked Sales by Product")
plt.xlabel("Quarters")
plt.ylabel("Sales")
plt.xticks(x, categories)
plt.legend()
plt.show()
````

### Example 20: Filled Area Plot

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = x
y2 = x**2 / 20

plt.fill_between(x, y1, y2, alpha=0.3, color='blue', label='Area between curves')
plt.plot(x, y1, label='y = x', linewidth=2)
plt.plot(x, y2, label='y = x²/20', linewidth=2)

plt.title("Filled Area Between Curves")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
````

### Example 22: Contour Plot

````python
import matplotlib.pyplot as plt
import numpy as np

# Create grid
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# Create 2D function (e.g., Z = X² + Y²)
Z = X**2 + Y**2

plt.contour(X, Y, Z, levels=15)
plt.colorbar(label='Value')
plt.title("Contour Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
````

## 7. 🚀 Advanced Concepts

### Advanced Features

**1. Object-Oriented Interface vs Pyplot Interface**

We've been using the simple `pyplot` interface (`plt.plot()`, etc.). Matplotlib also has an object-oriented interface that's more powerful for complex plots:

````python
import matplotlib.pyplot as plt

# Object-oriented approach (more control, better for complex plots)
fig, ax = plt.subplots(figsize=(8, 6))

x = [1, 2, 3, 4, 5]
y = [1, 4, 2, 3, 5]

ax.plot(x, y, marker='o', linewidth=2, label='Data')
ax.set_title("Object-Oriented Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.legend()
ax.grid(True)

plt.show()
````

**2. Custom Colormaps and Color Normalization**

````python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable

# Create data
x = np.random.randn(100)
y = np.random.randn(100)
values = x + y  # Values for coloring

# Create scatter plot with custom colormap
fig, ax = plt.subplots()
scatter = ax.scatter(x, y, c=values, cmap='coolwarm', s=100, alpha=0.6)
plt.colorbar(scatter, ax=ax, label='Value')
ax.set_title("Custom Colormap")
plt.show()
````

**3. GridSpec for Complex Layouts**

````python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(10, 8))
gs = gridspec.GridSpec(3, 3, figure=fig)

# Create subplots with different sizes
ax1 = fig.add_subplot(gs[0, :])  # First row, all columns
ax2 = fig.add_subplot(gs[1, :-1])  # Second row, first two columns
ax3 = fig.add_subplot(gs[1:, -1])  # Last two rows, last column
ax4 = fig.add_subplot(gs[-1, 0])  # Last row, first column
ax5 = fig.add_subplot(gs[-1, -2])  # Last row, second-to-last column

# Add data to each subplot
ax1.plot([1, 2, 3, 4], label='ax1')
ax2.scatter([1, 2, 3], [1, 2, 3], label='ax2')
ax3.bar([1, 2, 3], [1, 2, 3], label='ax3')
ax4.plot([1, 2, 3], label='ax4')
ax5.scatter([1, 2, 3], [1, 2, 3], label='ax5')

# Add legends
for ax in [ax1, ax2, ax3, ax4, ax5]:
    ax.legend()

plt.tight_layout()
plt.show()
````

### Best Practices

- ✅ Use the **object-oriented interface** for complex plots with multiple customizations
- ✅ Always add **labels and legends** so viewers understand what they're seeing
- ✅ Use **appropriate plot types** for your data (scatter for relationships, bars for categories, lines for trends)
- ✅ Keep colors **accessible** (consider colorblind-friendly palettes)
- ✅ Save high-quality plots with **dpi=300** or higher for publications
- ✅ Use **consistent fonts and sizes** across related plots

### Performance Tips

| Tip | Why it Helps |
|-----|---------------|
| Use `rasterized=True` for huge datasets | Rasterized plots are smaller files and render faster |
| Limit number of points displayed | Too many points slow down rendering and make plots cluttered |
| Use `fmt` parameter in `errorbar()` instead of separate plot calls | More efficient for multiple overlapping plots |
| Save as PNG for web, PDF for print | Different formats are optimized for different purposes |

### Common Mistakes Beginners Make

- ❌ Forgetting to call `plt.show()` (nothing appears!)
- ❌ Not adding labels to axes (viewers don't know what data is shown)
- ❌ Using default matplotlib colors (blue, orange, green) which can look unprofessional
- ❌ Creating huge plots with millions of points (very slow and hard to see patterns)
- ❌ Overlapping subplots without `tight_layout()` (text overlaps)
- ❌ Not using `alpha` for overlapping plots (can't see what's underneath)
- ❌ Mixing pyplot and object-oriented interfaces (confusing and error-prone)

### Debugging Tips

- 🔍 Check the **type** of data you're plotting (array vs list vs DataFrame)
- 🔍 Use `plt.figure()` to create a new plot when code doesn't show anything
- 🔍 Print your data before plotting to check its shape and values
- 🔍 Use `ax.grid(True)` to see data more clearly
- 🔍 Start simple (one line), then add customizations one by one

### 🟢 Beginner Project: Visualizing Test Scores

Create a dashboard showing test score analysis for a class.

````python
import matplotlib.pyplot as plt
import numpy as np

scores = np.array([78, 85, 92, 66, 74, 88, 95, 59, 81, 90])

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Histogram
axes[0, 0].hist(scores, bins=7, color='skyblue', edgecolor='black')
axes[0, 0].set_title("Score Distribution")
axes[0, 0].set_xlabel("Score")
axes[0, 0].set_ylabel("Frequency")
axes[0, 0].axvline(np.mean(scores), color='red', linestyle='--', label=f'Mean: {np.mean(scores):.1f}')
axes[0, 0].legend()

# Plot 2: Bar chart of individual scores
axes[0, 1].bar(range(len(scores)), scores, color='green', alpha=0.7)
axes[0, 1].set_title("Individual Student Scores")
axes[0, 1].set_xlabel("Student ID")
axes[0, 1].set_ylabel("Score")
axes[0, 1].axhline(y=70, color='red', linestyle='--', alpha=0.5, label='Passing Grade (70)')
axes[0, 1].legend()

# Plot 3: Pie chart (Pass/Fail)
passed = np.sum(scores >= 70)
failed = np.sum(scores < 70)
axes[1, 0].pie([passed, failed], labels=['Passed', 'Failed'], autopct='%1.1f%%', colors=['green', 'red'])
axes[1, 0].set_title("Pass/Fail Distribution")

# Plot 4: Box plot
axes[1, 1].boxplot(scores, vert=True)
axes[1, 1].set_ylabel("Score")
axes[1, 1].set_title("Score Statistics (Box Plot)")
axes[1, 1].grid(axis='y', alpha=0.3)

plt.suptitle("Class Test Score Dashboard", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
````

### 🟡 Intermediate Project: Stock Price Analysis

Visualize stock price trends over time with moving average.

````python
import matplotlib.pyplot as plt
import numpy as np

# Simulate stock price data
days = np.arange(1, 101)
base_price = 100
price = base_price + np.cumsum(np.random.randn(100) * 2)  # Random walk

# Calculate moving average (smoothed trend)
window = 10
moving_avg = np.convolve(price, np.ones(window)/window, mode='valid')

fig, ax = plt.subplots(figsize=(12, 6))

# Plot actual price
ax.plot(days, price, label='Stock Price', color='blue', linewidth=1.5, alpha=0.7)

# Plot moving average
ax.plot(days[window-1:], moving_avg, label=f'{window}-Day Moving Average', color='red', linewidth=2)

# Add fill between max and min
ax.fill_between(days, price, alpha=0.2, color='blue')

ax.set_title("Stock Price Analysis", fontsize=14, fontweight='bold')
ax.set_xlabel("Days")
ax.set_ylabel("Price ($)")
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

## 11. 🎤 Interview Questions with Answers

### Beginner-Level Questions

**Q1: What is Matplotlib?**
> A: Matplotlib is a Python library for creating static, animated, and interactive visualizations. It's used to plot data in the form of charts, graphs, and images.

**Q2: What is the difference between `plt` and `ax`?**
> A: `plt` (pyplot) is a module that provides a simple interface for basic plotting. `ax` (axes) gives you more control over individual plot elements when using the object-oriented interface.

**Q3: What does `plt.show()` do?**
> A: It displays the current plot in a window. Without it, the plot won't appear on screen (though it will save if you use `savefig()` before calling `show()`).

**Q4: How do you save a plot?**
> A: Using `plt.savefig("filename.png", dpi=300, bbox_inches='tight')`. This saves the current figure to a file.

**Q5: What is a `figure` and what is an `axes`?**
> A: A figure is the overall window/canvas. Axes are the actual plot areas inside the figure where data is drawn. You can have multiple axes in one figure.

### Intermediate Questions

**Q6: What's the difference between `plot()` and `scatter()`?**
> A: `plot()` connects points with lines (good for trends). `scatter()` shows just the points as dots (good for relationships and distributions).

**Q7: What does `alpha` parameter do?**
> A: `alpha` sets transparency of plot elements (0 = invisible, 1 = fully opaque). Useful for overlapping plots.

**Q8: What is a colormap and why is it useful?**
> A: A colormap is a function that maps data values to colors. It's useful for showing a third dimension of data in scatter/heat plots without a 3rd axis.

**Q9: How do you create subplots?**
> A: Using `plt.subplots(nrows, ncols)` which returns a figure and axes array. Then plot on each axes separately.

**Q10: What's the difference between `tight_layout()` and manual spacing?**
> A: `tight_layout()` automatically prevents subplots from overlapping. Manual spacing gives you more control but requires more code.

### Advanced Interview Questions

**Q11: Explain the difference between the pyplot interface and the object-oriented interface.**
> A: Pyplot interface (`plt.plot()`) is simple but less powerful for complex plots. OO interface (`ax.plot()`) gives you more control, better for multi-plot figures, and is more Pythonic.

**Q12: How would you optimize a plot with millions of data points?**
> A: Use `rasterized=True` to convert vector graphics to raster (saves space), reduce points using sampling, or use scatter plot instead of plot (faster rendering).

**Q13: What are the common backends in Matplotlib and when would you use each?**
> A: Backends are rendering systems. "Agg" (default) for saving files, "TkAgg" for GUI windows, "inline" for Jupyter notebooks. Choose based on where you want output.

**Q14: How do you create a plot that will work in both Jupyter notebooks and scripts?**
> A: Use `%matplotlib inline` in Jupyter, or include `plt.show()` which works in both environments.

**Q15: Explain broadcasting in context of Matplotlib color arrays.**
> A: You can pass a 1D array of values as the `c` parameter to `scatter()`, and Matplotlib will automatically scale it to colors using a colormap. Broadcasting makes this elegant.
